```python
import pandas as pd
import dask.dataframe as dd

import datetime as dt

from pathlib import Path
```

```python
df = pd.read_csv('../../Downloads/selection_under_randomization_file.csv')
filepaths_found = list()
filepaths_notfound = list()
files_not_found = pd.DataFrame()
```

```python
sites = (df['Site name'].unique())
```

```python
# for site_name in sites:
site_name = 'Carp Pond'
site = site_name.split()[0]
site_df = df.loc[df['Site name'] == site_name]

detection_files_site = list(pd.DatetimeIndex(pd.to_datetime(site_df['Datetime UTC'])).strftime(f'**/bd2__{site}_%Y%m%d_%H%M%S.csv'))

for file in detection_files_site:
    files_in_dir = list(Path(f'{Path.home()}/duty-cycle-investigation/data/raw/{site}').glob(pattern=file))
    if len(files_in_dir) > 0:
        file_as_path = Path(files_in_dir[0])
        # site_df = site_df[pd.to_datetime(site_df['Datetime UTC'])!=dt.datetime.strptime(file_as_path.name, f'bd2__{site}_%Y%m%d_%H%M%S.csv')]
        filepaths_found.append(file_as_path)
    else:
        filepaths_notfound.append(file)

files_not_found = pd.concat([files_not_found, site_df])
```
